```python

import requests
from bs4 import BeautifulSoup
import pandas as pd
import urllib.request
import bs4
import collections
import shutil
import os
import matplotlib.pyplot as plt
```

```python
liste_pokemon = ["pikachu", "bulbasaur", "charmander", "squirtle"]
```

```python

url = "https://pokemondb.net/pokedex/national"

response = requests.get(url)
response.raise_for_status()
soup = BeautifulSoup(response.content, "lxml")
print(soup.title.string)

pokemon_names = soup.findAll('a', class_='ent-name')
for name in pokemon_names:
    print(name.text)
```

```python

def get_page(pokemon_name):
    url = f"http://pokemondb.net/pokedex/{pokemon_name}"
    response = requests.get(url)
    response.raise_for_status()
    return BeautifulSoup(response.content, "lxml")

def caracteristiques(pokemon_name):
    soup = get_page(pokemon_name)
    data = collections.defaultdict()
    
    for table in soup.findAll('table', {'class': "vitals-table"})[0:4]:
        table_body = table.find('tbody')
        for rows in table_body.findChildren(['tr']):
            if len(rows) > 1:
                column = rows.findChild('th').getText()
                cells = rows.findChild('td').getText()
                cells = cells.replace('\t', '').replace('\n', ' ')
                data[column] = cells
                data['name'] = pokemon_name
    return dict(data)
```

```python

items = []
for e, pokemon in enumerate(liste_pokemon):
    print(e, pokemon)
    item = caracteristiques(pokemon)
    items.append(item)
    if e > 5:
        break

df = pd.DataFrame(items)
df.head()
```

```python

for e, pokemon in enumerate(liste_pokemon):
    print(e, pokemon)
    url = "https://img.pokemondb.net/artwork/{}.jpg".format(pokemon)
    response = requests.get(url, stream=True)

    with open('{}.jpg'.format(pokemon), 'wb') as out_file:
        shutil.copyfileobj(response.raw, out_file)
    if e > 4:
        break
```
